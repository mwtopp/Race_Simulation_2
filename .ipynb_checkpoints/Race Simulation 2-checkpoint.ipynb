This is the second project to look at the problem of simulating a motor race. As before the aim of the simulation is to determine the optimal lap (or laps) for a car to stop for a pit stop to replace tyres.

In this project I will develop the simulation further, specifically:

- I will use an object oriented method rather than the solely process oriented method of the previous simulation,

- The simulation will include multiple cars.

The assumptions of this model are:

- The aim is to complete the race in the highest placed position possible (if there is more than 1 car).

- If there is only one car, completing the race in the shortest possible time is the goal.

- Lap time is affected by the underlying speed of the car and driver, the current fuel load and the current tyre condition.

- Slower cars can be overtaken if there is a sufficient difference in speed / lap time.

- Car speed / lap time will be calculated at 1 second intervals and the distance travelled calculated accordingly.

For simplicity the model simulates the continuous process of a race using a discrete process. The actions of each car in the race are determined in turn, starting with the lead car and proceeding in order of race position, for each time interval. The time interval used in this model is one second. For increased accuracy smaller intervals could be used.

The process of overtaking is simplified to avoid adding a lot of complication to the model. Whereas in motor racing there is generally specific points on a track where it is possible for a car to overtake another competitive car, this model treats all points on the track equally. Also there is no allowance for the possibility that the driver in front makes a mistake or that the driver behind performs a brilliant overtake, the relative performance of the driver / car combinations is deterministic with no random element. The ease of making a successful overtake is governed by the overtake_req parameter which defines the minimum speed advantage required.

The variables are:

pit_lap = the lap on which the car will stop for a pit stop

The race parameters are:

lap_distance = the length of a single lap (in metres)

race_laps = the number of laps in a race

no_comp = the number of competitor cars in the race (in addition to the car of interest / team car)

pit_stop_time = the time that a pit stop takes (in seconds)

car_length = the minimum distance a following car can be behind another car (in metres)

overtake_req = the minimum advantage required by a car to overtake a slower car (in metres per second)

The car parameters are:

nominal_lap_time = the expected lap time for a car allowing for no fuel load and no tyre wear

max_fuel_penalty = the proportional increase in lap time for a car with a full fuel load

max_tyre_life = a notional maximum tyre life in laps, defined as the distance at which lap time is doubled

```python
# import the required libraries
from datetime import timedelta
import pandas as pd
from matplotlib import pyplot as plt
import random
```

```python
# create a car Class
class Car:
    
    # define an initialize function
    # include instance name
    # include instance nominal_lap_time
    # include instance max_fuel_penalty
    # include instance max_tyre_life
    # include instance pit_lap
    # set lap_count to 0
    # set stopped to False
    # set in_pit to False
    # set stop_duration to 0
    # set distance to 0
    # set distance_delta to 0
    # set tyre_wear to 0
    # set tyre_deg to 0
    # set tyre_per to 1
    # set fuel_factor to 1 + max_fuel_penalty
    # set lap_time to nominal_lap_time
    
    def __init__(self, name, nominal_lap_time, max_fuel_penalty, max_tyre_life, pit_lap):
        self.name = name
        self.nominal_lap_time = nominal_lap_time
        self.max_fuel_penalty = max_fuel_penalty
        self.max_tyre_life = max_tyre_life
        self.pit_lap = pit_lap
        self.lap_count = 0
        self.stopped = False
        self.in_pit = False
        self.stop_duration = 0
        self.distance = 0
        self.distance_delta = 0 ## race function defines distance_delta after each calc_lap_time call
        self.tyre_wear = 0
        self.tyre_deg = 0
        self.tyre_per = 1
        self.fuel_factor = 1 + max_fuel_penalty
        self.lap_time = nominal_lap_time
    
    # define a repr function
    def __repr__(self):
        return self.name
    
    # define a calculate lap_time method
    def calc_lap_time(self):
        self.lap_time = self.nominal_lap_time * self.fuel_factor * self.tyre_per
    
    # define a calculate fuel_factor method
    def calc_fuel_factor(self, race_laps, lap_distance):
        race_distance = race_laps * lap_distance
        self.fuel_factor = 1 + (((race_distance - self.distance) / race_distance) * self.max_fuel_penalty)
    
    # define a tyre_per method
    def calc_tyre_per(self, lap_distance):
        
        # calculate tyre_wear 
        self.tyre_wear += self.distance_delta * self.fuel_factor
        
        # calculate tyre_deg
        self.tyre_deg = self.tyre_wear / (self.max_tyre_life * lap_distance)
        
        # calculate tyre_per
        self.tyre_per = self.tyre_deg**2 + 1
        
    
```

```python
# create a function to create the team car (car of interest), set up the car_list and car_df
def create_team_car(team_name, nominal_lap_time, max_fuel_penalty, max_tyre_life, pit_lap):
    
    # create the team car
    team_car = Car(team_name, nominal_lap_time, max_fuel_penalty, max_tyre_life, pit_lap)
    
    # create car_list
    car_list = []

    # append to car_list
    car_list.append(team_car)

    # create list of car attributes
    team_attr = [[team_car.name, team_car.nominal_lap_time, team_car.max_fuel_penalty, 
                  team_car.max_tyre_life, team_car.pit_lap]]

    # create car_df
    car_df = pd.DataFrame(team_attr, columns = ['Car', 'Nom_Lap_Time', 'Max_Fuel_Pen', 'Max_Tyre_Life', 'Pit_Lap'])
    
    # return the car_list and car_df
    return car_list, car_df
```

```python
# define a function to generate the competitor cars and add them to the car_list and car_df
def create_competitors(car_list, car_df, no_comp, mean_nom_lap, std_nom_lap, mean_max_fuel, std_max_fuel,
                       mean_max_tyre, dif_max_tyre, pit_lap_list):
    
    # create comp_list
    comp_list = []
    
    # create a car instance for each no_comp
    for i in range(no_comp):
        
        # set nominal_lap_time
        nominal_lap_time = round(random.normalvariate(mean_nom_lap, std_nom_lap), 1)
        
        # set max_fuel_factor
        max_fuel_penalty = round(random.normalvariate(mean_max_fuel, std_max_fuel), 2)
        
        # set max_tyre_life
        max_tyre_life = random.randint(mean_max_tyre - dif_max_tyre, mean_max_tyre + dif_max_tyre)
        
        # set pit_lap
        pit_lap = random.choice(pit_lap_list)
        
        # create a Car instance
        comp_list.append(Car('Car {}'.format(i), nominal_lap_time, max_fuel_penalty, max_tyre_life, pit_lap))
        
    # add comp_list to car_list
    car_list = car_list + comp_list
    
    # iterate through cars in comp_list
    for i in range(len(comp_list)):
    
        # create list of car attributes
        car = comp_list[i]
        car_attr = [car.name, car.nominal_lap_time, car.max_fuel_penalty, car.max_tyre_life, car.pit_lap]
    
        # append car_attr to car_df
        car_df.loc[i + 1] = car_attr

    # return car_list and car_df
    return car_list, car_df
```

```python
# define a qualifying function that creates the start_grid_dict and returns the updated car_df
def qualifying(car_list, car_df, qual_lap_std, grid_offset):
    
    # create qual_lap_dict
    qual_lap_dict = {}
    
    # iterate through cars in car_list
    for car in car_list:
        
        # generate qualifying lap time
        qual_lap_dict[random.normalvariate(car.nominal_lap_time, qual_lap_std)] = car
        
    # create start_grid_dict
    start_grid_dict = {}
    
    # order qualifying lap times
    qual_lap_order = sorted(list(qual_lap_dict.keys()))
    
    # iterate through cars
    for i in range(len(qual_lap_order)):
    
        # order cars by qualifying lap time
        car = qual_lap_dict.get(qual_lap_order[i])
        
        # assign starting grid position
        start_grid_dict[i] = car
        
        # set car.distance according to grid position
        car.distance = -i * grid_offset
        
    # update car_df with start positions
    # add start_position column to car_df
    car_df['start_position'] = 0

    # iterate through grid positions
    for i in range(len(start_grid_dict)):
    
        # assign grid position to car
        car_df.loc[(car_df.Car == start_grid_dict[i].name), 'start_position'] = i
    
    # return start_grid_dict and car_df
    return start_grid_dict, car_df
```

```python
# define a function to run the race
def race(lap_distance, race_laps, start_grid_dict, car_df, pit_stop_time, car_length, overtake_req, race_time_limit, debug):
    
    # set race_finished to False
    race_finished = False
    
    # store car positions
    positions_dict = start_grid_dict
    
    # create finish_position column in car_df
    car_df['finish_position'] = 0
    
    # create finish_time column in car_df
    car_df['finish_time'] = 0
    
    # create positions_df with start_grid_dict
    ## create lists of cars and start positions
    car_list = list(start_grid_dict.values())
    start_position_list = list(start_grid_dict.keys())

    ## create a list of column headers
    column_list = ['Lap'] + car_list
    
    ## create the positions_df
    positions_df = pd.DataFrame(columns = column_list)
    
    ## populate the Laps column of positions_df
    positions_df['Lap'] = list(range(race_laps + 1))
    
    ## populate the lap 0 row of positions_df
    for n in range(len(car_list)):
        positions_df.loc[0, car_list[n]] = start_position_list[n]
    
    # store cars that have finished the race
    finished_cars = []
    
    # set race_time
    race_time = timedelta(seconds = 0)
    
    # check if time_limit is required
    if race_time_limit == True:
        
        # set time_limit
        time_limit = timedelta(seconds = race_laps * car_df['Nom_Lap_Time'].max() * 1.25)
        
    else:
        # set time_limit to arbitary value
        time_limit = timedelta(seconds = 0)
        
                
    # run race until all cars are finished or time_limit is reached
    while (race_finished == False) & ((race_time < time_limit) | (race_time_limit == False)):
    
        # check if all cars have finished the race
        if len(finished_cars) == len(car_list):
        
            # race complete return result
            race_finished = True
            
    
        # if not continue race
        else:
    
            # cycle through the cars in order of position
            for i in range(len(car_list)):
        
                # set car variable as the current car
                car = positions_dict[i]
                
                # debug option
                if debug >= 3:
                    print('{}: {} is current car in position {}.'.format(str(race_time), car, i))
                  
                # check if car has finished race
                if car in finished_cars:
                
                    # continue to next car 
                    continue
            
                # check if car completed a lap at last iteration
                if car.distance - (car.lap_count * lap_distance) > lap_distance:
                        
                    # update car.lap_count
                    car.lap_count += 1
                    
                    # add position to positions_df
                    positions_df.loc[car.lap_count, car] = i
                                           
                    # debug option
                    if debug >= 2:
                        print('{}: {} has completed lap {} in position {}'.format(str(race_time), car, car.lap_count, i))
                
                # check if car finished race at last iteration
                if car.distance > lap_distance * race_laps:
                
                    # add car to finished cars
                    finished_cars.append(car)
                    
                    # add finishing position to car_df
                    car_df.loc[(car_df.Car == car.name), 'finish_position'] = i
                    
                    # add finishing race_time to car_df
                    car_df.loc[(car_df.Car == car.name), 'finish_time'] = race_time
                    
                    # debug option
                    if debug >= 1:
                        print('{}: {} has been added to finished_cars'.format(str(race_time), car))
                    
                    # continue to next car
                    continue
            
                # check if car is in pits
                if car.in_pit == True:
                    
                    # add 1 to stop_duration
                    car.stop_duration += 1
                             
                    # check if car has completed pit stop
                    if car.stop_duration == pit_stop_time:
                    
                        # car leaves the pits
                        car.in_pit = False
                        
                        car.tyre_wear = 0
                        car.tyre_deg = 0
                        car.tyre_per = 1
                        
                        # debug option
                        if debug >= 1:
                            print('{}: {} has left the pits'.format(str(race_time), car))
                    
                        # set car.stopped to True and continue to next car
                        car.stopped = True
                        continue
                        
                    # continue to the next car
                    else:
                        continue
            
                # calculate distance travelled this iteration
                else:
                
                    # calculate current fuel_factor
                    car.calc_fuel_factor(race_laps, lap_distance)
                
                    # calculate current tyre_per
                    car.calc_tyre_per(lap_distance)
                
                    # calculate current lap_time
                    car.calc_lap_time()
                
                    # calculate maximum distance car could travel
                    max_dist = lap_distance / car.lap_time
                
                    # check if there is another car within the max_dist
                    # iterate through cars in front starting with the car immediately in front
                    # if car is not leading
                    if i == 0:
                        
                        # set distance_delta to the max_dist
                        car.distance_delta = max_dist
                        
                    else:
                        
                        # iterate through cars in front (starting with one position ahead)
                        for x in range((-i + 1), 1):
                        
                            # set car_in_front variable
                            car_in_front = positions_dict[-x]
                        
                            # check if car is behind car_in_front 
                            if (car.distance + max_dist) < (car_in_front.distance - car_length):
                            
                                # set distance_delta to the max_dist
                                car.distance_delta = max_dist
                                
                                # leave car_in_front for loop
                                break
                            
                            # else car.distance + max_dist is greater than car_in_front.distance
                            else:
                        
                                # check if car_in_front has finished race
                                if car_in_front in finished_cars:
                                    
                                    # set distance_delta to the max_dist
                                    car.distance_delta = max_dist
                                    
                                    # leave car_in_front for loop
                                    break
                            
                                # check if car_in_front is in the pits
                                elif car_in_front.in_pit == True:
                                
                                    # set distance_delta to the max_dist
                                    car.distance_delta = max_dist
                                
                                    # change positions_dict
                                    positions_dict[-x] = car
                                    positions_dict[-x+1] = car_in_front
                                
                                    # debug option
                                    if debug >= 2:
                                        print('{}: {} has overtaken {} while in pit'.format(
                                            str(race_time), car, car_in_front))
                        
                                # check if car has overtaken car_in_front 
                                elif (car.distance + max_dist) >= (car_in_front.distance + overtake_req):
                            
                                    # set distance_delta to the max_dist
                                    car.distance_delta = max_dist
                            
                                    # change positions_dict
                                    positions_dict[-x] = car
                                    positions_dict[-x+1] = car_in_front
                            
                                    # debug option
                                    if debug >= 2:
                                        print('{}: {} has overtaken {}'.format(str(race_time), car, car_in_front))
                            
                                # else car is stuck behind car_in_front
                                else:
                            
                                    # set distance_delta to a restricted value
                                    car.distance_delta = car_in_front.distance - car.distance - car_length
                            
                                    # debug option
                                    if debug >= 2:
                                        print('{}: {} is stuck behind {}'.format(str(race_time), car, car_in_front))
                                        
                                    # leave car_in_front for loop
                                    break
                                   
                    
                    # set updated distance
                    car.distance += car.distance_delta
                    
                    # debug option
                    if debug >= 3:
                        print('{}: {} has travelled {}'.format(str(race_time), car, car.distance_delta))
                    
                    # debug option
                    if debug >= 3:
                        print('{}: {} distance is {}'.format(str(race_time), car, car.distance))
                                         
                    # check if car is due for a pit stop
                    # assume the pit entry is 50m before the start / finish line
                    if (car.distance > ((car.pit_lap * lap_distance) - 50)) & (car.stopped == False):
                
                        # car enters pits
                        car.in_pit = True
                    
                        # debug option
                        if debug >= 1:
                            print('{}: {} has entered the pits'.format(str(race_time), car))
                            
        # update race_time
        race_time += timedelta(seconds=1)
    
    # return positions_df and car_df
    return positions_df, car_df                
```

The aim of the simulation is to determine the optimal lap for the team car to make it's pit stop for a particular race. For the following implementation of the simulation the race parameters will be (based roughly on BTCC cars and Knockhill, although there are no scheduled pit stops in BTCC):

lap_distance = 2000 metres

race_laps = 40

no_cars = 20

pit_stop_time = 25 seconds

car_length = 1 metres

overtake_req = 2 metres per second

The mean car parameters will be:

nominal_lap_time = 50 seconds

max_fuel_penalty = 0.1

max_tyre_life = 50

To simulate the race we need to predict when the competitors will stop for their pit stop. A reasonable approach is to assume that the competitors will choose to stop on the lap that results in the shortest time to complete the race. We will also assume that our competitors do not have access to a race simulation as detailed as this and will choose a pit lap based on a single car on track (similar to my first race simulation model).

Hence we will initially determine the optimal pit lap with no competitor cars on track.

```python
# define the parameters
lap_distance = 2000
race_laps = 40
pit_stop_time = 25
car_length = 1
overtake_req = 2
nominal_lap_time = 50
max_fuel_penalty = 0.1
max_tyre_life = 50

# create result_df_list
result_df_list = []

# iterate through possible pit laps (excluding first lap and last lap)
for i in range(race_laps - 1):
    
    # define the car pit_lap variable
    pit_lap = i + 1
    
    # create the team_car (assuming mean values for car parameters)
    car_list, car_df = create_team_car('Python Racing', nominal_lap_time, 
                                       max_fuel_penalty, max_tyre_life, pit_lap)
    
    # create the start_grid_dict and car_df
    start_grid_dict, car_df = qualifying(car_list, car_df, qual_lap_std = 0.1, grid_offset = 2)
    
    # run the race
    result_positions, result_df = race(lap_distance, race_laps, start_grid_dict,
                car_df, pit_stop_time, car_length, overtake_req, race_time_limit=False, debug=0)
    
    # append result_df to result_df_list
    result_df_list.append(result_df)
    
# concatenate result_df
sim_result_df = pd.concat(result_df_list, ignore_index=True)
```

```python
# plot the race times
finish_times = [t.total_seconds() for t in sim_result_df['finish_time']]

plt.figure(figsize=(8, 6))
plt.plot(sim_result_df['Pit_Lap'], finish_times, 'o')

plt.title('Race Finish Time against Pit Lap')
plt.xlabel('Pit Lap')
plt.ylabel('Finish Time (seconds)')

plt.show()
```

We can see that stopping on either lap 19 or 20 is optimal. Stopping on lap 18 or 21 is two seconds slower.

Assuming our competitors will use an optimal strategy, they will stop either on lap 19 or 20.

So we can now simulate the race including the competitors.

To reduce the calculations required we will initially limit the possible pit_laps to laps 10 through 30 as it is very unlikely, judging from the previous simulation, that stopping before lap 10 or after lap 30 will be a successful strategy.

The optimal strategy may vary depending on the expected performance of the team car, for example what if our car is faster than the average car, what if it has better tyre duration than the average car? For the first simulation we will assume our car is average in nominal_lap_time, max_fuel_penalty and max_tyre_life.

As we repeat the race simulation with different pit_lap values we want the competitor cars to remain the same so the results are comparable. Therefore we will set the random seed at the start of each iteration.

Before we start lets take a look at the randomly generated competitors we will be racing against.

```python
# investigate the competitors
# define parameters for simulation
nominal_lap_time = 50
mean_nom_lap = 50
max_fuel_penalty = 0.1
max_tyre_life = 50
mean_max_tyre = 50

# set the pit_lap
pit_lap = 19

# set the number of competitor cars
no_comp = 19

# set the random seed
random.seed(6)

# create the team_car
car_list, car_df = create_team_car('Python Racing', nominal_lap_time, max_fuel_penalty, max_tyre_life, pit_lap)

# create the competitors
car_list, car_df = create_competitors(car_list=car_list, car_df=car_df, no_comp=no_comp, mean_nom_lap=mean_nom_lap,
                                      std_nom_lap=1, mean_max_fuel=max_fuel_penalty, std_max_fuel=0, 
                                      mean_max_tyre=mean_max_tyre, dif_max_tyre=5, pit_lap_list=[19, 20])
    
# create the start_grid_dict and car_df
start_grid_dict, car_df = qualifying(car_list, car_df, qual_lap_std = 0.1, grid_offset = 2)
    
# view the car_df
car_df
```

Now we can proceed with the first simulation.

```python
# define parameters for simulation
lap_distance = 2000
race_laps = 40
pit_stop_time = 25
car_length = 1
overtake_req = 2
nominal_lap_time = 50
mean_nom_lap = 50
max_fuel_penalty = 0.1
max_tyre_life = 50

# set the number of competitor cars
no_comp = 19

# create result_df_list
result_df_list = []

# iterate through possible pit laps
for i in range(10, 31):
    
    # set the random seed to ensure each iteration is comparable
    random.seed(6)
    
    # define the car pit_lap variable
    pit_lap = i
    
    # create the team_car (assuming mean values for car parameters)
    car_list, car_df = create_team_car('Python Racing', nominal_lap_time, 
                                       max_fuel_penalty, max_tyre_life, pit_lap)
    
    # create the competitors
    car_list, car_df = create_competitors(car_list=car_list, car_df=car_df, no_comp=no_comp, mean_nom_lap=mean_nom_lap,
                                      std_nom_lap=1, mean_max_fuel=max_fuel_penalty, std_max_fuel=0, 
                                      mean_max_tyre=max_tyre_life, dif_max_tyre=5, pit_lap_list=[19, 20])
    
    # create the start_grid_dict and car_df
    start_grid_dict, car_df = qualifying(car_list, car_df, qual_lap_std = 0.1, grid_offset = 2)
    
    # run the race
    result_positions, result_df = race(lap_distance, race_laps, start_grid_dict,
                car_df, pit_stop_time, car_length, overtake_req, race_time_limit=False, debug=0)
    
    # append result_df to result_df_list
    result_df_list.append(result_df.iloc[[0]])
    
# concatenate result_df
sim_result_df = pd.concat(result_df_list, ignore_index=True)
```

Next we will investigate the optimal pit strategy for a car faster than the average car. We will set the nominal_lap_time one standard deviation faster than the mean.

```python
# define parameters for simulation
lap_distance = 2000
race_laps = 40
pit_stop_time = 25
car_length = 1
overtake_req = 2
nominal_lap_time = 49
mean_nom_lap = 50
max_fuel_penalty = 0.1
max_tyre_life = 50

# set the number of competitor cars
no_comp = 19

# create result_df_list
result_df_list = []

# iterate through possible pit laps
for i in range(10, 31):
    
    # set the random seed to ensure each iteration is comparable
    random.seed(6)
    
    # define the car pit_lap variable
    pit_lap = i
    
    # create the team_car
    car_list, car_df = create_team_car('Python Racing', nominal_lap_time, 
                                       max_fuel_penalty, max_tyre_life, pit_lap)
    
    # create the competitors
    car_list, car_df = create_competitors(car_list=car_list, car_df=car_df, no_comp=no_comp, mean_nom_lap=mean_nom_lap,
                                      std_nom_lap=1, mean_max_fuel=max_fuel_penalty, std_max_fuel=0, 
                                      mean_max_tyre=max_tyre_life, dif_max_tyre=5, pit_lap_list=[19, 20])
    
    # create the start_grid_dict and car_df
    start_grid_dict, car_df = qualifying(car_list, car_df, qual_lap_std = 0.1, grid_offset = 2)
    
    # run the race
    result_positions, result_df = race(lap_distance, race_laps, start_grid_dict,
                car_df, pit_stop_time, car_length, overtake_req, race_time_limit=False, debug=0)
    
    # append result_df to result_df_list
    result_df_list.append(result_df.iloc[[0]])
    
# concatenate result_df
sim_result_df = pd.concat(result_df_list, ignore_index=True)
```

```python
# create a function to plot the car positions
def plot_positions(result_df, result_positions):

    # create a figure
    plt.figure(figsize=(16, 7))
    ax = plt.subplot()

    # iterate through cars
    for i in range(len(result_df)):
    
        # define car
        car = result_positions.columns[i + 1]
    
        # draw line showing position of car by lap
        plt.plot(result_positions['Lap'], result_positions[car], label=car)

    # set labels, ticks etc
    plt.xticks(list(range(len(result_positions))), list(range(len(result_positions))))
    ax.invert_yaxis()
    plt.yticks(list(range(len(result_df))), list(range(len(result_df))))
    plt.xlabel('Lap')
    plt.ylabel('Race Position')
    plt.title('Car Position by Lap')
    plt.legend()

    plt.show()
```

Notes on the model.

After successfully testing the simulation with a few competitors, some strange results occured and the race function could slip into an infinite loop when running it with large numbers of competitors. On investigation it became apparent that with twenty cars on track behaviours that had seemed unlikely actually happened quite frequently. In particular in a queue of cars the car at the back was able to gain distance because the restricted distance was set by the car two or more places in front overriding the restricted distance set by the car immediately in front. This could cause a car to overtake two other cars at once and in doing so one car could be removed from the positions_dict. Consequently the race would never finish because the length of the list of finished cars would not match the number of cars. Once diagnosed it was easy to solve by adding a break after the first time a car was stuck behind a car in front.

Possible developments to model.

The option to stop more than once or not at all.

Different tyre types could be introduced, ie. high grip / low durabilty or low grip / high durability.

Tracks could be simulated in greater detail, ie. ease of overtaking at each section of track could be included.
